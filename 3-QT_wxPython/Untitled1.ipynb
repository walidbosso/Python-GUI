```python
import wx
import wx.grid
import sqlite3

sqliteConnection = sqlite3.connect('pharmacy.db')
cursor = sqliteConnection.cursor()
print("Successfully Connected to SQLite")


class Client:
    def __init__(self, client_id, first_name, last_name, number):
        self.client_id = client_id
        self.first_name = first_name
        self.last_name = last_name
        self.number = number


class Employee:
    def __init__(self, employee_id, name, salary, phone):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary
        self.phone = phone


class Product:
    def __init__(self, product_id, name, description, price, category):
        self.product_id = product_id
        self.name = name
        self.description = description
        self.price = price
        self.category = category


class Purchase:
    def __init__(self, purchase_id, customer_id, product_id, employee_id, price, quantity, date):
        self.purchase_id = purchase_id
        self.customer_id = customer_id
        self.product_id = product_id
        self.employee_id = employee_id
        self.price = price
        self.quantity = quantity
        self.date = date


class MainUI(wx.Frame):
    def __init__(self):
        super(MainUI, self).__init__(None, title="Pharmacy Management", size=(800, 600))

        self.panel = wx.Panel(self)
        self.notebook = wx.Notebook(self.panel)

        self.client_tab = ClientPanel(self.notebook)
        self.employee_tab = EmployeePanel(self.notebook)
        self.product_tab = ProductPanel(self.notebook)
        self.purchase_tab = PurchasePanel(self.notebook)

        self.notebook.AddPage(self.client_tab, "Clients")
        self.notebook.AddPage(self.employee_tab, "Employees")
        self.notebook.AddPage(self.product_tab, "Products")
        self.notebook.AddPage(self.purchase_tab, "Purchases")

        sizer = wx.BoxSizer(wx.VERTICAL)
        sizer.Add(self.notebook, 1, wx.EXPAND)
        self.panel.SetSizer(sizer)

        self.Show()


class ClientPanel(wx.Panel):
    def __init__(self, parent):
        super(ClientPanel, self).__init__(parent)
        self.ClientsTable = wx.grid.Grid(self)
        self.ClientsTable.CreateGrid(0, 4)
        self.ClientsTable.SetColLabelValue(0, "Client ID")
        self.ClientsTable.SetColLabelValue(1, "First Name")
        self.ClientsTable.SetColLabelValue(2, "Last Name")
        self.ClientsTable.SetColLabelValue(3, "Number")

        self.client_Id = wx.TextCtrl(self, style=wx.TE_READONLY)
        self.client_firstName = wx.TextCtrl(self)
        self.client_lastname = wx.TextCtrl(self)
        self.client_number = wx.TextCtrl(self)

        self.edit_button = wx.Button(self, label="Edit Selected Client")
        self.edit_button.Bind(wx.EVT_BUTTON, self.edit_selected_client)

        self.create_button = wx.Button(self, label="Create Client")
        self.create_button.Bind(wx.EVT_BUTTON, self.createClient)

        self.delete_button = wx.Button(self, label="Delete Client")
        self.delete_button.Bind(wx.EVT_BUTTON, self.delete_client)

        self.setup_layout()

    def setup_layout(self):
        sizer = wx.BoxSizer(wx.VERTICAL)

        sizer.Add(self.ClientsTable, 1, wx.EXPAND | wx.ALL, 5)

        sizer.Add(wx.StaticText(self, label="Client ID:"), 0, wx.ALL, 5)
        sizer.Add(self.client_Id, 0, wx.EXPAND | wx.ALL, 5)

        sizer.Add(wx.StaticText(self, label="First Name:"), 0, wx.ALL, 5)
        sizer.Add(self.client_firstName, 0, wx.EXPAND | wx.ALL, 5)

        sizer.Add(wx.StaticText(self, label="Last Name:"), 0, wx.ALL, 5)
        sizer.Add(self.client_lastname, 0, wx.EXPAND | wx.ALL, 5)

        sizer.Add(wx.StaticText(self, label="Number:"), 0, wx.ALL, 5)
        sizer.Add(self.client_number, 0, wx.EXPAND | wx.ALL, 5)

        sizer.Add(self.edit_button, 0, wx.EXPAND | wx.ALL, 5)
        sizer.Add(self.create_button, 0, wx.EXPAND | wx.ALL, 5)
        sizer.Add(self.delete_button, 0, wx.EXPAND | wx.ALL, 5)

        self.SetSizer(sizer)

    def edit_selected_client(self, event):
        table = self.ClientsTable
        selected_items = table.GetSelectedCells()
        if selected_items:
            selected_row = selected_items[0][0]
            selected_values = [table.GetCellValue(selected_row, col) for col in range(4)]

            self.client_Id.SetValue(str(selected_values[0]))
            self.client_firstName.SetValue(selected_values[1])
            self.client_lastname.SetValue(selected_values[2])
            self.client_number.SetValue(selected_values[3])


if __name__ == "__main__":
    MainUI().MainLoop()
```
